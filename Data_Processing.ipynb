#2 Player Data


```python
with open ('Player_Stats.json') as f:
  data = json.load(f)

df_players = pd.DataFrame(data[0]['player_stats'])
df_players.head()
```

Finding Special Characters and Cleaning Names

```python
import re

player_names = df_players["player"]

special_characters = []

# Iterate through each player's name
for player_name in player_names:
    special_chars = re.findall(r'[^\w\s]', player_name)  # Find special characters
    special_characters.extend(special_chars)  # Add special characters to the list

# Remove duplicates and sort the list
special_characters = sorted(list(set(special_characters)))

print("List of special characters:", special_characters)
```

Cleaning

```python
df_players['player'] = df_players['player'].replace(replacement_dict, regex=True)
df_players['team'] = df_players['team'].replace(replacement_dict, regex=True)
```

```python
df_players.to_csv('Player_Stats.csv', index = False)
```
